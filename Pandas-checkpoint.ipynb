## Making a pivot table

```python
import pandas as pd
import os
```

```python
pop = pd.read_csv('Population_total.csv')
```

```python
pop
```

### .pivot() returns DF organized by index/col values (without aggregation)

```python
#Dropping null values form DF, using .dropna() method
pop.dropna()

#write inplace to save changes to the file or just reassign the changed values to the original df
#new_pop = pop.dropna()
pop.dropna(inplace=True)
```

```python
pop
```

```python
#pivot requires 3 arguments INDEX,COLUMNS,VALUES
pop_pivot = pop.pivot(index='year',columns='country',values='population')
```

```python
pop_pivot
```

```python
#now that pivot table is made we can call what we want from that table
pop_pivot[['India','China']]
```

```python
sel_pop = pop_pivot[['India','China','Brazil','Afghanistan','Pakistan']]
```

```python
sel_pop
```

# Data Visualizations using pandas

## Lineplot

```python
#One necessary argument .plot() method takes in 'kind'
#figsize() is a tuple used to resize x axis y axis respectively
sel_pop.plot(kind='line',xlabel='Years',ylabel='Population in billion',title='Population from 1955-2020',figsize=(8,4))

sel_pop.to_excel('pivot_table.xlsx')
```

## Single bar plot

```python
#selecting only one year
sel_pop[sel_pop.index.isin([2020])]
```

```python
pop_2020 = sel_pop[sel_pop.index.isin([2020])]
```

```python
#formatting by .T method to transpose the data 
pop_2020= pop_2020.T
```

```python
pop_2020
```

```python
pop_2020.plot(kind='bar',color='purple',xlabel='Countires',ylabel='Population in billion',title='Population 2020')
```
